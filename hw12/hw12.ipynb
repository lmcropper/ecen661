# Illuminance **E** vs Luminance **L**

**Goal:** know what each quantity means, what it measures, and how to compute it.

## Where these fit (photometry quick map)

| What? | Symbol | “Per…” | Unit |
|---|---:|---|---|
| **Luminous flux** (total light) | Φ | — | lumen ($lm$) |
| **Luminous intensity** (directional) | I | per solid angle | candela ($cd = lm/sr$) |
| **Illuminance** (incident on a surface) | E | per receiving area | lux ($lx = lm/m^2$) |
| **Luminance** (apparent brightness of a source) | L | per projected area per solid angle | $cd/m^2$ |

## Lux = Illuminance (how much light lands here)

**Definition**  
$$E = \frac{d\Phi}{dA}$$

- $\Phi$: lumens arriving on the surface  
- $A$: receiving area

**Units:** $1\ \text{lux} = 1\ \frac{\text{lumen}}{\text{m}^2}$

**Intuition:** lux is about the receiver (desk, sensor, floor), not the lamp.

## From candelas to lux (inverse-square)

For a point source aimed at the surface (normal incidence):

$I=\frac{d \Phi}{d \Omega}$

$I \, d \Omega = d \Phi$

The definition of solid angle is $d \Omega =\frac{dA}{r^2}$

$I \, \frac{dA}{r^2} = d \Phi$

$\frac{I}{r^2}=\frac{d \Phi}{dA}=E$

$E = \frac{I}{r^2}$

More generally, if the surface normal is at angle $\theta$ to the incoming ray:
$E = \frac{I\cos\theta}{r^2}$

This is called the inverse square law.

**HW5:** For easy reading a classroom should have an illuminance of 500 lux in the center of the room at the desk.  The ceiling is made with uniformly space L1HZ-30801 White LEDs.  What is the spacing of the LEDs in units of m? Treat the LEDs as point sources. Assume that the distance from the ceiling to the desk is 3m.

**HELP:** The intensity is $I=42 \cos \theta$.

Directly under one LED would be $E_1=\frac{42}{9}=4.7$.  There is a grid of LEDs in both the $x$ and $y$ directions.  The distance from the LED to the desk is $r= \sqrt{z^2+(i*\Delta)^2+(j*\Delta)^2}$.  The illuminance is then given by $E_{i,j}=\frac{42}{r^2}*(\cos \theta)^2$.  One $\cos \theta$ is for the projection angle and one $\cos \theta$ is for LED emission pattern.


```python
# HW5: find LED grid spacing that yields 500 lux at the desk
target_illuminance = 500.0  # lux
height = 3.0  # meters, ceiling to desk
on_axis_intensity = 42.0  # candela for each LED
max_index = 50  # +/- LEDs accounted for in each direction

indices = np.arange(-max_index, max_index + 1)
XX, YY = np.meshgrid(indices, indices, indexing="ij")


def grid_illuminance(delta):
    """Sum the illuminance contribution from a 2-D LED lattice with spacing delta."""
    r_sq = height**2 + delta**2 * (XX**2 + YY**2)
    return np.sum(on_axis_intensity * height**2 / r_sq**2)


def find_spacing():
    low, high = 0.05, 3.0  # initial bracket in meters
    E_low = grid_illuminance(low)
    while E_low < target_illuminance:
        low *= 0.5
        E_low = grid_illuminance(low)

    E_high = grid_illuminance(high)
    while E_high > target_illuminance:
        high *= 1.5
        E_high = grid_illuminance(high)

    for _ in range(80):  # binary search until convergence
        mid = 0.5 * (low + high)
        E_mid = grid_illuminance(mid)
        if abs(E_mid - target_illuminance) < 1e-3:
            return mid, E_mid
        if E_mid > target_illuminance:
            low = mid
        else:
            high = mid
    mid = 0.5 * (low + high)
    return mid, grid_illuminance(mid)


hw5, achieved_E = find_spacing()
print(
    "Spacing Δ = {:.3f} m gives {:.1f} lux at the desk.".format(hw5, achieved_E)
)
```

## Power from One Surface to Another
When we have an extended source we need to characterize the power from one surface to another.  

<img src="extended.jpg" width="500">

The key quantity used to describe the amount of light emitted in a specific direction from the extended source is the radiance (or luminance) as given by

$$L = \frac{d^2 \Phi}{d\Omega \, dA_{\perp}}$$
The perpendicular area is the projection resulting in 
$$L = \frac{d^2 \Phi}{d\Omega \, dA \cos \theta}$$
You can also convert the radiance/luminance to be related to the intensity resulting in 
$$L = \left(\frac{d \Phi}{d\Omega} \right) \left( \frac{d \Phi}{dA \cos \theta} \right) = \frac{dI}{dA \cos \theta} $$

If the radiance/luminance is constant with angle it is called a Lambertian source and is given by
$$I(\theta)=I_o \cos \theta$$
$$L(\theta)=L_o$$
Lambertian sources are ideal diffuse or matte surfaces like projector screens, wall, roads, where light is scattered evenly.

Let's call the source surface $S$.  An infitesmal area is designated as $dS$.
<img src="extended3.jpg" width="500">

The luminosity term is

$$L=\frac{d^2 \Phi}{d \Omega \, dS \, \cos \theta}$$

$d \Omega$ is the infinitesmal solid angle from the source point to the area on the receive plane.  $d \Omega$ is defined as

$$d \Omega = \frac{dA \, \cos \phi}{R^2}$$

Solve for $\Phi$ to get

$$d \Phi^2 = L \, d \Omega \, dS \cos \theta$$

Replace $d \Omega$ to get

$$d \Phi^2 = L \frac{dS \cos \theta \, dA \cos \phi}{R^2}$$

Integrate over the source and receiver surfaces to get the power at the receiver.

$$\Phi = L \int_{S} \int_{A}\frac{dA \cos \psi}{R^2}\, dS \cos \phi$$

## Two Planar Surface

Let's look at setting up a problem for two planar surfaces facing each other with a Lambertian source.

<img src="extended_example1.jpg" width="500">

Since the surfaces are parallel to each other, the two angles are equal $\theta=\phi \equiv \psi$.

$$\Phi=L\int_{S_0}\int_{A_0}\frac{da_0 \cos \psi}{R^2}\, dS_0 \cos \psi$$

$L$ is outside of the integral because $L$ is constant for a Lambertian source.

Distance relation:

$$R^2 = Z^2 + (X_1 - X_2)^2 + (Y_1 - Y_2)^2$$



Geometry relation:

$$(\cos \psi)^2=\frac{(X_1 - X_2)^2 + (Y_1 - Y_2)^2}{(X_1 - X_2)^2 + (Y_1 - Y_2)^2 + Z^2}$$

Full integral form:

$$\Phi =\int_{S_0}\int_{A_0}L\frac{(X_1 - X_2)^2 + (Y_1 - Y_2)^2}{\left[(X_1 - X_2)^2 + (Y_1 - Y_2)^2 + Z^2\right]^2}\, dx_1 \, dy_1 \, dx_2 \, dy_2$$

This is a hard integral to set up and solve.  So, we want to analyze two simipler systems:
- **CASE 1:** $x_1, y_1 \ll R$
- **CASE 2:** $x_1, y_1 \gg R$

## Case 1: Distant Source

Rather than analyzing the power at the receiver, let's look at the illuminance, $E$,  at the receiver.  

$$E = \frac{\Phi}{A}$$

The differential illuminance is 

$$dE = \frac{d^2 \Phi}{d a_o}= L \frac{\cos \theta \cos \psi}{R^2} dS_o$$

This eliminates the integration over the receiver, resulting in 

$$E =\int_{S_o} L \frac{\cos \theta \cos \psi}{R^2} dS_o$$

The approximation is that $x_1, y_1 \ll R$.  This means that $S_o \ll R^2$, which makes $R$, $\cos \theta$, and $\cos \phi$ approximately constant relative to the integal.  Thus, they can be taken out of the intergral resulting in

$$E =  L \frac{\cos \theta \cos \psi}{R^2} \int_{S_o} dS_o$$

Performing the integration results in 

$$E =  L \frac{\cos \theta \cos \psi \, S_o}{R^2}$$

## Case 2: Large Source

Consider a circular Lambertian source of radius $R$.  We wish to determine the illuminance on a receiving surface $dA$ perpendicular to the axis, a distance $z$ away.

<img src="large.jpg" width="500">

Solid angle from the source point to the differential source disc is
$$d \Omega=\frac{\text{area}}{\text{squared radius}}=\frac{dA \, \cos \theta}{(z/\cos \theta)^2}=\frac{dA}{z^2}\cos^3 \theta$$
$$d^2\Phi= L \, dS \, \cos \theta d \Omega= L \, dS \cos \theta \frac{dA}{z^2} \cos^3 \theta$$
$$\frac{d^2 \Phi}{dA}=\frac{L}{z^2}\cos^4 \theta \,dS$$

In polar coordinates $dS=r\,dr\,d\phi$ resulting in 
$$\frac{d^2 \Phi}{dA}=\frac{L}{z^2}\cos^4 \theta \,r\,dr\,d\phi$$
Now use $r=z \, \tan \theta$.  Taking the dervative of $r$ results in $dr=(z/ \cos^2 \theta) d\theta$ resulting in

$$\frac{d^2 \Phi}{dA}=\frac{L}{z^2}\cos^4 \theta \, z\,\tan \theta \, \frac{z}{\cos^2 \theta}  \,d\phi$$

$$\frac{d^2 \Phi}{dA}=L\cos^2 \theta \, \tan \theta \, d\theta \, d\phi$$
$$=L\, \cos \theta \, \sin \theta \, d\theta \, d\phi$$
$$=\frac{1}{2} L\, \sin 2\theta \, d\theta \, d\phi$$

$$E=\int \left( \frac{d^2\Phi}{dA}\right)=\frac{1}{2} \int_0^{2\pi} d\phi \int_0^{\theta_o}L \sin 2 \theta \, d\theta$$

If the source is Lambertian
$$E=\frac{1}{2}(L)(2\pi) \int_0^{\theta_o} \sin 2 \theta \, d\theta$$
$$E=\pi \, L \left( -\frac{1}{2} \cos 2 \theta \big|_0^{\theta_o} \right)$$
$$E=\pi \, L \, \frac{1}{2} \left( - \cos 2 \theta_o + 1 \right)$$
$$E=\pi \, L \, \frac{1}{2} \left( 2 sin^2 \theta_o \right)$$
$$\boxed{E=\pi \, L \, sin^2 \theta_o}$$

```python
# HW15: illuminance at a person 1 m in front of the transmitting Lambertian screen
luminous_power = 5700.0  # lm emitted by LED
led_diameter = 22e-3  # m
led_radius = led_diameter / 2
A_led = np.pi * led_radius**2

# Optical layout
d1 = 0.42
f1 = 0.36
d12 = 5.04
f2 = 2.5
lens_diameter = 0.10
screen_transmission = 0.70
viewer_distance = 1.0  # m in front of the screen

# Image geometry at the screen
si1 = 1 / (1 / f1 - 1 / d1)
do2 = d12 - si1
si2 = 1 / (1 / f2 - 1 / do2)
m_total = (-si1 / d1) * (-si2 / do2)
screen_radius = led_radius * abs(m_total)
A_screen_image = A_led * m_total**2

# Flux delivered to the screen by lens 1 (Lambertian source approximation)
aperture_area = np.pi * (lens_diameter / 2) ** 2
solid_angle_lens1 = aperture_area / d1**2
flux_to_screen = luminous_power * solid_angle_lens1 / np.pi

# Screen transmitted exitance and luminance (Lambertian transmission)
E_screen_incident = flux_to_screen / A_screen_image
E_screen_transmitted = screen_transmission * E_screen_incident
L_screen = E_screen_transmitted / np.pi  # cd/m^2

# Illuminance at viewer from finite circular Lambertian source
theta0 = np.arctan(screen_radius / viewer_distance)
hw15 = np.pi * L_screen * np.sin(theta0) ** 2

print(f"Screen diameter = {2 * screen_radius:.3f} m")
print(f"Screen transmitted illuminance = {E_screen_transmitted:.6f} lux")
print(f"Viewer illuminance (hw15) = {hw15:.6f} lux")
```
