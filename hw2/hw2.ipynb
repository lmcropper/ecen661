# Overview and Review
Start with Maxwell's equations
$$\nabla \times \vec{\mathbf{E}} = -\frac{\partial \vec{B}} {\partial t}$$
$$\nabla \times \vec{H} =  J + \frac{\partial \vec{Dv}} {\partial t}$$
$$\nabla \cdot \vec{D} = \rho_v $$
$$\nabla \cdot \vec{B} = 0 $$

In optics we do not use metal wires to create sources so we have a charge free and 
curent source free media resulting in 
$$\rho_v=0 $$
$$\vec{J} = \sigma \vec{\mathbf{E}}.$$

The constitutive relations relate $\vec{\mathbf{E}}$ and $\vec{D}$  and $\vec{H}$ and $\vec{B}$ as given by
$$\vec{D} = \epsilon_o \vec{\mathbf{E}} + \vec{P} $$
$$\vec{H} = \frac{\vec{B}}{\mu_o} - \vec{M} $$

We often talk about LIH media (linear, isotropic, homogeneous).

# Linear Media
In general $\vec{P}$ depends on the magnitude of $\vec{\mathbf{E}}$ as given by
$$P = \epsilon_o \chi E + 2d E^2 +4 \chi^{(3)} E^3 + \cdot \cdot \cdot$$

If the higher order terms are insignificant then the material is linear

$$P = \epsilon_o \chi E $$

$$\vec{D} = \epsilon_o E + \epsilon_o \chi \vec{\mathbf{E}}$$
$$= \epsilon_o \left( 1+ \chi \right) \vec{\mathbf{E}} $$
$$= \epsilon_o \epsilon_r \vec{\mathbf{E}} $$

The field of nonlinear optics will not be covered in this class.

Applications: electro-optic modulators, wavelength converters

# Isotropic Media
In general $\epsilon_r$ is a tensor (matrix) rather than a constant.  The permittivity depends on the direction of the electric field.  

Propagation in these anisotropic materials will not be covered in this class.

Applications: polarization rotators, liquid crystal.


# Homogeneous Material
A homogeneous material has no spatial variations in the material.  All material has spatial variation 
for example at the atomic level there is a nucleaus and electron clouds, etc.  The homogeneity is only 
important relative to the wavelength.  If the spatial variations are small compared to the wavelength then
the material can be assumed as being effectively homogenous.

If the media is completely homogeneous then there will be very little control over the field.  For example 
a lens is a spatial variation and thus a lens is a non-homogenous material.  So we will be 
analyzing propagation through non-homogeous materials but we will look at propagation through homogeneous sections.

# Scalar Wave Equation
What we want to accomplish is to calculate the phase and amplitude of an optical wave at an arbitrary position 
after it has traveled through a set of optical components.  The components can be apertures, lenses, gratings, 
etc.

The basic idea is to calculate the electric field at a particular plane given a known field at a different plane.

First of all, we assume that the fields are time harmonic as given by
$$E(x,y,z,t) = \Re \left\{ A(x,y,z) e^{j \omega t} \right\}$$

This assumes that the field only has a single wavelength. A wave that has a single wavelength is called monochromatic.  
The analysis of non monochromatic waves is called coherence theory.  This topic will not be covered in this class.

When describing the optical fields the complex form is often used with the $\Re \{ \}$ and the $e^{j \omega t}$ parts assumed.  

The time harmonic form of Maxwell's equations are given by
$$\nabla \times \vec{\mathbf{E}} = -j \omega \mu \vec{H}$$
$$  \nabla \times \vec{H} = \sigma \vec{\mathbf{E}} + j \omega \epsilon \vec{\mathbf{E}}$$
$$ = (j \omega \epsilon + \sigma) \vec{\mathbf{E}}$$
$$ = j \omega \epsilon_c \vec{\mathbf{E}}$$
$$ \nabla \cdot \vec{D} = 0$$
$$ \nabla \cdot \vec{B} = 0$$


Combine the two curl equations to create one second order differential equation rather than two coupled
first order differential equations.
$$\nabla \times \left( \nabla \times \vec{\mathbf{E}} \right) = \nabla \times \left( -j \omega \mu \vec{H} \right) $$
$$ = - j \omega \mu \left( \nabla \times \vec{H} \right)$$

**What assumption was made at this point?**
$$\nabla \times \left( -j \omega \mu \vec{H} \right) = - j \omega \mu \left( \nabla \times \vec{H} \right)$$
This means that **$\mu$ is independent of the spatial coordinatate ($x$, $y$, and $z$)**.


$\mu$ does not have any spatial variation.  This is valid since we are dealing with non-magnetic materials ($\mu_R=1$).

Use the equation $ \nabla \times \vec{H} = j \omega \epsilon_c \vec{\mathbf{E}}$
$$\nabla \times \left( \nabla \times \vec{\mathbf{E}} \right)  = -j \omega \mu \left( j \omega \epsilon \vec{\mathbf{E}} \right)$$

Now we use the vector identity
$$\nabla \times \nabla \times \vec{\mathbf{E}} = \nabla \left( \nabla \cdot \vec{\mathbf{E}} \right) - \nabla^2 \vec{\mathbf{E}}$$

**The next approximation is linear media**
$$\nabla \cdot \vec{\mathbf{E}} = \nabla \cdot \left( \frac{\vec{D}}{\epsilon} \right)$$

For a **homogeneous material** $\epsilon$ is independent of the spatial coordinates ($x$, $y$, and $z$).
$$= \frac{1}{\epsilon} \left( \nabla \cdot \vec{D} \right)$$
$$ = 0 $$

The wave equations becomes 
$$\nabla \times \left( \nabla \times \vec{\mathbf{E}} \right) = -j \omega \mu \left( j \omega \epsilon \vec{\mathbf{E}} \right)$$
$$0-\nabla^2 = \omega^2 \mu \epsilon \vec{\mathbf{E}}$$
$$\nabla^2 \vec{\mathbf{E}} + k^2 \vec{\mathbf{E}} = 0$$

# Geometrical Optics
In geometrical (or ray) optics we use the approximation that the variation of the material is much slower than the spatial derivatives to show that we can model structures by tracing rays perpendicular to the wavefront.

Start with an arbitrary field given by
$$\vec{\mathbf{E}}(r) = \vec{\mathbf{E}}_o(r) \exp \left[ -j k_o \phi(r) \right]$$
and
$$\vec{\mathbf{H}}(r) = \vec{\mathbf{H}}_o(r) \exp \left[ -j k_o \phi(r) \right]$$

The term $k_o \phi$ defines the planes of constant phase (the wavefront).  The term $k_o$ is a constant and $\phi$ is a function of ($x$, $y$, and $z$).

Apply Maxwell's equation
$$\nabla \times \vec{\mathbf{E}} = -j \omega \mu \vec{\mathbf{H}}$$
The electric field has both a vector $\left( \vec{E}_o \right)$ and a scalar component $\left( e^{-j k_o \phi} \right).$

Use the vector identity
$$\nabla \times \left( f \vec{\mathbf{A}} \right) = f \left( \nabla \times \vec{\mathbf{A}} \right) + \left( \nabla f \right) \times \vec{\mathbf{A}},$$
to get 
$$\nabla \times \vec{\mathbf{E}} = e^{-j k_o \phi} \left( \nabla \times \vec{\mathbf{E}}_o \right) + \nabla \left\{ e^{-j k_o \phi}\right\} \times \vec{\mathbf{E}}_o$$

# Basic Postulates of Geometrical Optics
(1) Rays are normal to the wavefront and vice versa.

(2) Rays satisfy the laws of reflection and refraction (Snell's law).

(3) The optical path length along any ray between two wavefronts are equal.

(4) Irradiance at any point is proportional to the ray density at that point.



Other resulting corollaries are:

(1) In a uniform medium, light travels along straight lines.

(2) The optical paths are reversible.

(3) The optical path difference between two neighboring rays is zero.


Ray tracing can be used to:

(1) Analyze optical systmes

(2) Illumination analysis/desig

(3) Image artifacts

# Prisms

Optical prisms are versatile elements made from transparent materials like various types of glass, designed to refract or reflect light.  Main applications (1) alter direction, and (2) dispersing.

## Deviation

![D](deflection.png)

This figure shows all of the angles of the beam through a prism.  $\alpha_i$ are the incident angles and $\beta_i$ are the refracted angle (defined by Snell's law). 

The dashed line from A to B is the line if there was no prism. $\delta_1$ is the deflection at the first surface.

The dashed line from C to B is the direction of the final beam.  $\delta_2$ is the deflection at the second surface. 

$\delta$ is the angle from the undeviated beam path to the actual beam path.  So, the total deflection is $\delta=\delta_1+\delta_2$.

 At the first surface $\alpha_1=\delta_1+\beta_1$, which can be rearranged to give $\delta_1=\alpha_1-\beta_1$.

At the second surface $\beta_2=\delta_2+\alpha_2$, which can be rearranged to give $\delta_2=\beta_2-\alpha_2$.

Thus, $\delta=\delta_1+\delta_2=\alpha_1+\beta_2-(\beta_1 +\alpha_2)$.

$\beta_1$ and $\alpha_2$ are defined relative to the lines drawn normal to the prism surface, resulting in 

$\beta_1+\alpha_2+(180-\sigma$)=180 or $\beta_1+\alpha_2=\sigma$.

$\delta=\alpha_1+\beta_2-\sigma$

**HW7:** Use an equilateral prism ($\sigma=60$) with a refractive index of $n_{pr}=1.5$. Create a plot of the beam deviation $\delta$ for $30<\alpha_1<80$.

```python
# === PARAMETERS (edit these as needed) ===
n0 = 1.00  # outside medium (air)
npr = 1.50  # prism refractive index
sigma_deg = 60.0  # apex angle of prism, degrees

def prism_deviation(alpha1_deg, n0, npr, sigma_deg):
    """
    Compute total deviation delta for a single incidence angle alpha1 (degrees).
    Returns None if total internal reflection occurs.
    Run everything in radians internally.
    """
    alpha1 = np.pi/180*alpha1_deg
    sigma=np.pi/180*sigma_deg

    # Snell at first surface: n0 * sin(alpha1) = npr * sin(beta1)
    s = (n0 / npr) * np.sin(alpha1)
    if max(abs(s)) > 1.0:
        print("first surface TIR")
        return None  # total internal reflection
    beta1 = np.asin(s)

    # Geometry inside prism
    alpha2 = sigma - beta1

    # Snell at second surface: np * sin(b2) = n0 * sin(a2)
    s2 = (npr / n0) * np.sin(alpha2)
    if max(abs(s2)) > 1.0:
        print("second surface TIR")
        print("alpha2 (deg) =", 180/np.pi*alpha2)
        return None  # total internal reflection at second surface
    beta2 = np.asin(s2)

    # Total deviation
    delta = alpha1 + beta2 - sigma  # radians
    return 180/np.pi*delta

# === SWEEP -10 < alpha1 < 10 ===
alpha1=np.linspace(30,80,10)
delta=prism_deviation(alpha1, n0, npr, sigma_deg)
plt.plot(alpha1, delta)
plt.title("HW2_7: Prism Deviation vs Incidence Angle", color='red')
plt.xlabel("Incidence Angle (degrees)")
plt.ylabel("Deviation Angle (degrees)") 
plt.show()
```

## 1) Direction Cosines  Concept

A electromagnetic plane wave can be characterized by its propagation constant given by
$$\vec{k}=k_x \hat{x}+k_y \hat{y}+k_z\hat{z}$$
where 
$$\sqrt{k_x^2+k_y^2+k_z^2}=k_o=\frac{2 \pi}{\lambda}.$$

The propagation constant provides the energy flow direction.  The plane wave can be abstracted into a more arbitrary wave by replacing it with a set of rays that can be propagated through a system.  Rather than using the propagation vector we can replace the ray with a vector that indicates the propagation direction.  

Given a vector $\mathbf{A}$ in 3D that makes angles $\alpha$, $\beta$, and $\gamma$ with the $x$, $y$, and $z$ axes respectively, the **direction cosines** are

$$a = \cos\alpha, \quad b = \cos\beta, \quad c = \cos\gamma$$

If $\lVert \mathbf{A}\rVert$ is the magnitude and $(A_x,A_y,A_z)$ are its components, then

$$
A_x = \lVert \mathbf{A}\rVert\,a, \quad
A_y = \lVert \mathbf{A}\rVert\,b, \quad
A_z = \lVert \mathbf{A}\rVert\,c
$$

For a **unit vector** $\hat{\mathbf{u}}$, its components **are** its direction cosines:

$$\hat{\mathbf{u}} = (a,b,c)$$

A key identity:

$$a^2 + b^2 + c^2 = 1$$

which follows from $\lVert \hat{\mathbf{u}}\rVert^2 = 1$.

## 2) Fermat Principle  Concept
The tangential component of the wavevector $\vec{k}$ is equal across an interface.

Let's look at this for Snell's law.

**Example:** Let's look at a 2D example. A beam is traveling in the $xz$ plane. It is incident at an angle of $\theta_i$ (measured from the surface normal).  The surface is in $xy$ plane. The index of refraction of the incident region is $n_i$ and the index of refraction of the transmission region is $n_t$.  The wavevector of the incident wave is 

$\vec{k}=n_i k_o \left( sin \theta_i \hat{x} - cos \theta_i \hat{z} \right)$,

where $k_o=\frac{2 \pi}{\lambda}$

This wavevector has a magnitude of $n_i k_o$.

The tangential or parallel component is $k_{//}=n_i k_o sin \theta_i$.

In the transmitted region the tangential or parallel component is the equal and the magnitude of the wavevector is $n_t k_o$.  Thus, in the transmitted region the wavevector is

$\vec{k}_t=n_i k_o sin \theta_i \hat{x} +k_\perp \hat{z}$

Using the magnitude, we get

$\left( n_t k_o \right)^2 = \left( n_i k_o sin \theta_i \right)^2 + k^2_\perp$

Solving for the angle in the transmitted region results in

$sin \theta_t=\frac{k_{//}}{n_t k_o} = \frac{n_i k_o sin \theta_i }{n_t k_o} =  \frac{n_i sin \theta_i }{n_t}$

Rearranging this we get a common form of Snell's law

$n_t sin \theta_t= n_i sin \theta_i$







## 2) Vector Reflection & Refraction (Snell's Law in Vector Form)

For a smooth interface between two homogeneous, isotropic media with refractive indices $n_1$ (incident) and $n_2$ (transmission):

- $\hat{\mathbf{i}}$ = unit incident vector (toward the interface)  
- $\hat{\mathbf{n}}$ = unit surface normal (pointing from medium 1 into medium 2)  
- $\hat{\mathbf{r}}$, $\hat{\mathbf{t}}$ = reflected and transmitted (refracted) unit vectors

## 3) Reflection
Basic process:

(1) Divide the incident wave up into parallel and perpendicular components.

Find the projection of the incident ray onto the surface normal. $\hat{i}_\perp=\left( \hat{i} \cdot \hat{n} \right) \hat{n} $

Subtract the projection from the incident wave to get the portion of the incident wave parallel to the surface. 

$\hat{i}_{//}=\hat{i} - \left( \hat{i} \cdot \hat{n} \right) \hat{n} $

(2) The parallel component is the same in the incident and reflected wave.  The perpendicular component is opposite.

Subtract the projection again to get the reflected wave $ \boxed{\hat{r}=\hat{i} - 2 \left( \hat{i} \cdot \hat{n} \right) \hat{n} }$

## 4)  Refraction (Snells Law in Vector Form)
The basic concept is the following.  (1) Break incident wave up into components parallel and perpendicular to the interface (based on the surface normal).  (2) The parallel component of $\vec{k}$ is equal across the interface.  (3) Use Snell's law to calculate the perpendicular component.

$\hat{i} = \hat{i}_{//} + \hat{i}_\perp$

$\hat{k}_i = n_i k_o \left( \hat{i}_{//} + \hat{i}_\perp \right)$

$\hat{k}_t = n_t k_o \left( \hat{t}_{//} + \hat{t}_\perp \right)$

$n_t \hat{t} - n_i \hat{i} = n_t k_o \hat{t}_{//} + n_t k_o \hat{t}_\perp  +n_i k_o \hat{i}_{//}  +n_i k_o \hat{i}_\perp$ 

Since the parallel components are equal $n_t k_o \hat{t}_{//} = n_i k_o \hat{i}_{//}$,

$n_t \hat{t} - n_i \hat{i} = n_t k_o \hat{t}_\perp  + n_i k_o \hat{i}_\perp = F \hat{n}$.

$F$ is a scalar.

To find the value of $F$ we can take a dot product of the equation with respect to $\hat{n}$ and use the fact that $\hat{n}$ is a unit vector $\left( \hat{n} \cdot \hat{n} = 1\right)$.

This results in 

$\hat{n} \cdot \left( n_t \hat{t} - n_i \hat{i} \right) = F \hat{n} \cdot \hat{n}$, 

which is simplified to give

$n_t \left( \hat{t} \cdot  \hat{n} \right)  - n_i \left( \hat{i} \cdot  \hat{n} \right) = F $.

Since $\hat{i}$, $\hat{t}$ and $\hat{n}$ are all unit vectors that means that $\left( \hat{t} \cdot  \hat{n} \right) =-cos \theta_t$ and $\left( \hat{i} \cdot  \hat{n} \right) =cos \theta_i$.  Notice that  $\left( \hat{t} \cdot  \hat{n} \right)$ is negative.  This negative is beacuse the transmitted wave is traveling in the opposite direction from the surface normal.  

Resulting in 

$F=-n_t*cos \theta_t -n_i cos \theta_i$

Go back to the equation for $n_t \hat{t} - n_i \hat{i}$.  Solving this equation for $\hat{t}$ results in the following derivation:

Starting with

$n_t \hat{t} - n_i \hat{i} = F \hat{n}$.

divide by $n_t$

$\hat{t} - \frac{n_i}{n_t} \hat{i} = \frac{F}{n_t} \hat{n}$.

Solve for $\hat{t}$ to get

$\hat{t} = \frac{n_i}{n_t} \hat{i} + \frac{F}{n_t} \hat{n}$.

Plug in the value for $F$ to get

$\hat{t} = \frac{n_i}{n_t} \hat{i} + \frac{-n_t cos \theta_t -n_i cos \theta_i}{n_t} \hat{n}$, 

$ = \frac{n_i}{n_t} \hat{i} + \left( -cos \theta_t -\frac{n_i}{n_t} cos \theta_i \right) \hat{n}$.

We want to covert everything into unit vectors ($\hat{i}$ and $\hat{n}$) and dot products. 

(1) Start by converting $\theta_t$ into $\theta_i$.  We want to use Snell's law ($n_i sin \theta_i = n_t sin \theta_t$). 

$cos \theta_t = \sqrt{cos^2 \theta_t}$

using a trig identity

$=\sqrt{1-sin^2 \theta_t}$

using Snell's law

$=\sqrt{1-\left( \frac{n_i}{n_t} \right)^2 sin^2 \theta_i}$

convert the $sin$ back into a $cos$ using a trig identity

$cos \theta_t=\sqrt{1-\left( \frac{n_i}{n_t} \right)^2 \left( 1- cos^2 \theta_i \right)}$

Plug back into the equation for $\hat{t}$

$\hat{t} = \frac{n_i}{n_t} \hat{i} + \left( -cos \theta_t -\frac{n_i}{n_t} cos \theta_i \right) \hat{n}$.

$\hat{t} = \frac{n_i}{n_t} \hat{i} + \left( -\sqrt{1-\left( \frac{n_i}{n_t} \right)^2 \left( 1- cos^2 \theta_i \right)} -\frac{n_i}{n_t} cos \theta_i \right) \hat{n}$.

Replace the $cos$ with the unit vector dot product $cos \theta_i =\hat{i} \cdot \hat{n} $ to get

$\hat{t} = \frac{n_i}{n_t} \hat{i} + \left( -\sqrt{1-\left( \frac{n_i}{n_t} \right)^2 \left( 1- \left( \hat{i} \cdot \hat{n} \right)^2 \right)} -\frac{n_i}{n_t} \left( \hat{i} \cdot \hat{n} \right) \right) \hat{n}$.

$\hat{t} = \frac{n_i}{n_t} \hat{i} - \left(\frac{n_i}{n_t} \left( \hat{i} \cdot \hat{n} \right) + \sqrt{1-\left( \frac{n_i}{n_t} \right)^2 \left( 1- \left( \hat{i} \cdot \hat{n} \right)^2 \right)}\right) \hat{n}$.

  Sellmeiers formula provides a reliable approximation of the refractive index, $n(\lambda_0)$, as a function of the freespace wavelength $\lambda_0$.  This formula is given as $$n^2 \left( \lambda_0 \right) = 1 + \sum_{i=1}^3 \frac{A_i \lambda_0^2}{\lambda_0^2  B_i}$$

```python
#HW15 SOLUTION
#DELETE THIS CODE BLOCK WHEN SENDING TO STUDENTS
def index_ref(glass, wavelength):
    #Returns the index of refraction for given glass type and wavelength in nm.
    #Parameters:
    #glass (str): Type of glass ('BK7', 'FS', 'MP', 'SF11', 'MgF', 'SiN')
    # Sellmeier coefficients from Table 2A.1 ( in microns)
    if glass == 'BK7':
        A = np.array([1.03961212, 0.231792344, 1.01046945])
        B = np.array([6.00069867e-3, 2.00179144e-2, 103.560653])
    elif glass == 'FS':
        A = np.array([0.696166300, 0.407942600, 0.897479400])
        B = np.array([4.67914826e-3, 1.35120631e-2, 97.9340025])
    elif glass == 'MP':
        A = np.array([1.3182408, 0.0244, 1.08915181])
        B = np.array([8.79e-3, 6.09e-2, 110.0])
    elif glass == 'SF11':
        A = np.array([1.73759695, 0.313747346, 1.89878101])
        B = np.array([13.188707e-3, 6.23068142e-2, 155.23629])
    elif glass == 'MgF':
        A = np.array([0.48755108, 0.39875031, 2.3120353])
        B = np.array([1.8821784e-3, 0.89518885e-2, 566.13559])
    elif glass == 'SiN':
        A = np.array([3.0249, 40314.0, 0])
        B = np.array([1.8317078e-2, 1537208.18, 0])
    else:
        raise ValueError("Unknown glass type")

    n2 = 1+A[0]*wavelength**2/(wavelength**2 - B[0]) + A[1]*wavelength**2/(wavelength**2 - B[1]) + A[2]*wavelength**2/(wavelength**2 - B[2])
    return np.sqrt(n2)

print("refractive index for SF11 at 500nm: ", index_ref('SF11', 0.500))
```
