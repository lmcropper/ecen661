## Basic Fourier Transform Relations


If $f(x) \leftrightarrow F(f)$:

-   $K f(x) \leftrightarrow K F(f)$
-   $f(ax) \leftrightarrow \frac{1}{|a|} F\!\left(\frac{f}{a}\right)$
-   $f(x-x_0) \leftrightarrow F(f)e^{-j2\pi f x_0}$
-   $e^{j2\pi f_0 x} f(x) \leftrightarrow F(f-f_0)$

# Telescope Resolution

A telescope is designed to image distant objects. We want to look at the pattern at the focus of two closely spaced stars. The stars are modeled as point sources. Since the light is broadband (white light), the system is considered to be incoherent. This means that you add the irradiance rather than the electric fields.  

Let's start with the on-axis signal.

The electric field at the entrance aperture of the telescope is a plane wave given by

$\vec{E}_1=\vec{E}_o \, e^{jkz}$.

We set the entrance aperture of the telescope at $z=0$ resulting in the electric field being 

$\vec{E}_1=\vec{E}_o$.

For a circular pupil of diameter $D$, the electric field at the focus of the telescope is 

$E_1(f_r) = E_{01} \frac{2J_1 \left( \pi D f_r \right)}{\pi D f_r}$  with $f_r=\frac{r}{\lambda f}$

Resulting in 

$E_1(r) = E_{01} \frac{2J_1\!\left(\frac{\pi D r}{\lambda f}\right)}{\frac{\pi D r}{\lambda f}}$

The corresponding irradiance is

$I=I_0\left[\frac{2J_1\!\left(\frac{\pi D r}{\lambda f}\right)}{\frac{\pi D r}{\lambda f}}\right]^2$ 


Now consider the second point source that arrives at a small angle $\theta$ relative to the optical axis in the $x$-$z$ plane.  The incident plane wave can be written as

$\vec{E}_2=\vec{E}_{02} \, e^{jk(x\sin\theta+z\cos\theta)}$ 

At the entrance pupil of the telescope ($z=0$), this reduces to a linear phase ramp across the aperture,

$\vec{E}_2=\vec{E}_{02} \, e^{jkx\sin\theta} = \vec{E}_{02} \, e^{j 2 \pi \frac{\theta}{\lambda}x} $

The electric field at the focus is similar to the on-axis image except for the frequency shift.


$E_2(f_r) = E_{02} \frac{2 J_1 \left( \pi D \left( f_r - \frac{\theta}{\lambda} \right) \right)}{\left ( \pi D \left( f_r-\frac{\theta}{\lambda} \right) \right)}$  with $f_r=\frac{r}{\lambda f}$

Resulting in 

$E_2(r) = E_{02} \frac{2 J_1 \left( \pi D \left( \frac{r}{\lambda f} - \frac{\theta}{\lambda} \right) \right)}{\left ( \pi D \left( \frac{r}{\lambda f}-\frac{\theta}{\lambda} \right) \right)}$  

$E_2(r) = E_{02} \frac{2 J_1 \left( \frac{\pi D}{\lambda} \left( \frac{r}{f} - \theta \right) \right)}{\left ( \frac{\pi D}{\lambda} \left( \frac{r}{f}-\theta \right) \right)}$  

This electric field profile has a maximum at $r=\theta f$. This is the same result as using geometric optics.


# Rayleigh Criterion

Start with the electric field:

$E_2(r) = E_{02} \frac{2 J_1 \left( \frac{\pi D}{\lambda} \left( \frac{r}{f} - \theta \right) \right)}{\left ( \frac{\pi D}{\lambda} \left( \frac{r}{f}-\theta \right) \right)}$


The intensity is proportional to the square of the field:

$I = I_o \left[\frac{2 J_1(u)}{u}\right]^2$

where

$u = \frac{\pi D}{\lambda} \left( \frac{r}{f} - \theta \right)$

Two point sources are just resolved when the central maximum of one pattern coincides with the first zero of the other.


The first zero of the Bessel function is $J_1(u) = 0 \quad \text{at} \quad u =3.8317$

- Source 1 at $\theta = 0$
- Source 2 at $\theta = \Delta\theta =\frac{r}{f}$

Substitute into $u$ to get $u = \frac{\pi D}{\lambda} \Delta\theta$

$\frac{\pi D}{\lambda} \Delta\theta = 3.8317$

$\Delta\theta = \frac{3.8317}{\pi} \frac{\lambda}{D} \approx 1.22 \frac{\lambda}{D}$


Notice that the resolution does not depend on the focal length or f-number of the telescope, but only on the size of the telescope with respect to wavelength.

For the Rayleigh criteria the dip between the two peaks is defined as 

$\left(2 \frac{J_1 \left( 3.8317/2 \right)}{3.8317/2} \right)^2+\left(2 \frac{J_1 \left( 3.8317/2 \right)}{3.8317/2}\right)^2$

$=2\left(2 \frac{J_1 \left(1.916 \right)}{1.916} \right)^2$

```python
y=2*(2 * special.j1(3.8317/2) / (3.8317/2)) ** 2
print("Intensity at midpoint = ", np.round(y,4))
```
