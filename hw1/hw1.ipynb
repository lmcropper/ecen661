**HW1:**  The Fresnel reflection coefficients (https://en.wikipedia.org/wiki/Fresnel_equations) determine the fraction of light that is reflected off an interface. (1) Create a Markdown box and put the equation in the box for both s-polarized and p-polarized light.  (2) Create a red box around the equation block.

For S-polarized light:

$R_s = \left\lvert \frac{Z_2cos\theta_i - Z_1cos\theta_t}{Z_2cos\theta_i + Z_1cos\theta_t} \right\rvert^2$

For P-polarized light:

$R_p = \left\lvert \frac{Z_2cos\theta_t - Z_1cos\theta_i}{Z_2cos\theta_t + Z_1cos\theta_i} \right\rvert^2$

**HW2:** Create a plot of the reflection coefficient for s-polarized light as a function of the incident angle for an angle from -80 degees to +80 degrees if the incident region is air ($n_{inc}=1$) and the transmission region is glass with a refractive index of $n_{tr}=1.5$.  Make sure that it has a red title.

```python
# Calculate reflection coefficient for s-polarized light
theta_i_deg = np.linspace(-80, 80, 1000)
theta_i_rad = np.radians(theta_i_deg)

# Incident region is air (n=1), transmission region is glass (n=1.5)
n_inc = 1
n_tr = 1.5

# Calculate transmission angle using Snell's law
sin_theta_t = (n_inc / n_tr) * np.sin(theta_i_rad)
theta_t_rad = np.arcsin(sin_theta_t)

# Impedance of each medium
Z_inc = 1 / n_inc  # normalized
Z_tr = 1 / n_tr    # normalized

# Fresnel reflection coefficient for s-polarized light
R_s = np.abs((Z_tr * np.cos(theta_i_rad) - Z_inc * np.cos(theta_t_rad)) / 
             (Z_tr * np.cos(theta_i_rad) + Z_inc * np.cos(theta_t_rad))) ** 2

plt.figure(figsize=(10, 6))
plt.plot(theta_i_deg, R_s)
plt.xlabel('Incident Angle (degrees)')
plt.ylabel('Reflection Coefficient')
plt.ylim(0, 1)
plt.title('S-Polarized Light Reflection Coefficient', color='red')
plt.grid(True)
plt.show()
```

**HW3:** What is the reflection coefficient if the incident angle is 20 degrees and it has s-polarized light?  Use a variable name of hw1_3.

```python
query_rad = np.radians(20)
hw1_3 = np.interp(query_rad, theta_i_rad, R_s)
print(f"Reflection coefficient at 20 degrees: {hw1_3}")
```
