# Green's Theorem

Let $U(\mathbf{r})$ and $G(\mathbf{r})$ be scalar functions.

$$\iiint_V (U \nabla^2 G - G \nabla^2 U)\, dV
=
\iint_S \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS$$

Notice that Green's functions relates a surface integral to a volume integral.

The surface that we choose is an arbitrary surface with a sphere removed at the origin.  The sphere at the origin is because the spherical wave goes to infinity at the origin $\left( \frac{1}{r} \right)$.

<img src="origin.jpg" width="300">

Going back to Green's theorem, which is 

$$\iiint_V (U \nabla^2 G - G \nabla^2 U)\, dV
=
\iint_S \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS$$

We use Green's Theorem on the V' surface resulting in 

$$\iiint_{V'} (U \nabla^2 G - G \nabla^2 U)\, dV
=
\iint_{S'} \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS$$

$S'=S + S_{\epsilon}$

$U$ and $G$ are both electric fields. 


## Maxwell's Equations in a Source-Free Medium

<details>
  <summary>Click to show wave equation derivation</summary>

Start with Maxwell's equation in a source free media

$\nabla \times \mathbf{E} = -\mu \frac{\partial \mathbf{H}}{\partial t}$

$\nabla \times \mathbf{H} = \epsilon \frac{\partial \mathbf{E}}{\partial t}$

$\nabla \cdot \mathbf{D} = 0$

$\nabla \cdot \mathbf{B} = 0$

Take the curl of Faraday's law to get

$\nabla \times (\nabla \times \mathbf{E})= \nabla \times  \left\{-\mu \frac{\partial \mathbf{H}}{\partial t}\right\}$

Assuming homogeneous $\mu$ means that $\mu$ is independent of spatial components and thus independent of $\nabla$.  You can also perform the spatial derivatives before the time variable.  The results are

$\nabla \times (\nabla \times \mathbf{E})= -\mu \frac{\partial}{\partial t} \left\{ \nabla \times \mathbf{H} \right\}$

Substitute Ampere's law to get

$\nabla \times (\nabla \times \mathbf{E})= -\mu \frac{\partial}{\partial t} \left\{ \epsilon \frac{\partial \mathbf{E}}{\partial t} \right\}$

$\nabla \times \nabla \times \mathbf{E} = -\mu \epsilon \frac{\partial^2 \mathbf{E}}{\partial t^2}$

Using the vector identity $\left[ \nabla \times \nabla \times \mathbf{E} = \nabla (\nabla \cdot \mathbf{E}) - \nabla^2 \mathbf{E} \right]$ to get

$\nabla (\nabla \cdot \mathbf{E}) - \nabla^2 \mathbf{E} = -\mu \epsilon \frac{\partial^2 \mathbf{E}}{\partial t^2}$

Using Gauss's Law $\left( \nabla \cdot \mathbf{D} = 0 \right)$ and $\mathbf{D} = \epsilon \mathbf{E}$ results in 

$\nabla \cdot \epsilon \mathbf{E} = 0$ 

For a linear homogeneous medium $\epsilon$ is independent of spatial coordinates making it independent of $\nabla$

$\epsilon \nabla \cdot \mathbf{E} = 0$ or $\nabla \cdot \mathbf{E} = 0$

So the wave equation becomes

$\nabla^2 \mathbf{E} - \mu \epsilon \frac{\partial^2 \mathbf{E}}{\partial t^2} = 0$

Using time-harmonic fields, each time derivative results in an $j \omega$.  The wave equation becomes

$\nabla^2 \mathbf{E} - \mu \epsilon \left( j \omega \right)^2 \mathbf{E} = 0$

$\nabla^2 \mathbf{E}  + \mu \epsilon \omega^2 \mathbf{E} = 0$

$\nabla^2 \mathbf{E}  + k^2 \mathbf{E} = 0$
</details>

 Thus,

$\nabla ^2 G = -k^2 \, G$  and $\nabla ^2 U = -k^2 \, U$.

resulting in 

$$\iiint_{V'} (U (-k^2 \, G) - G (-k^2 \, U) \, dV=
\iint_{S'} \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS$$

re-order to get

$$-\iiint_{V'} (k^2 \, U \, G - k^2 \, U \, G) \, dV
=
\iint_{S'} \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS$$

The two terms are equal and opposite resulting in

$$0=\iint_{S'} \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS$$

Now split the surface integral into the two pieces, $S'=S+S_{\epsilon}$ ($S$ is the outer surface and $S_{\epsilon}$ is the surface surrounding the origin)

$$0=\iint_{S} \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS + \iint_{S_{\epsilon}} \left( U \frac{\partial G}{\partial n}- G \frac{\partial U}{\partial n} \right) dS $$

Now let's look at the two functions $G$ and $U$.  $U$ will be our electric field.  We can chose an arbitrary function for $G$.  

In choosing $G$ we look at Huygens' Principle.  

Huygens’ Principle (1678) says: "Every point on a wavefront acts as a source of new small wavelets, and the new wavefront is the surface tangent to all those wavelets."

## Aperture
Consider a uniform field over an aperture.

The surface $S$ consists of:

-   $S_1$: the aperture plane
-   $S_2$: a large hemisphere of radius $R$

<img src="aperture.jpg" width="300">

Start by looking at the surface $S_2$.  As $R \to \infty$

$G = \frac{e^{jkR}}{R} \to 0$

Resulting in the integral over the $S_2$ surface to be zero. 

The $S_1$ surface is opaque except over aperture $\Sigma$

Assumptions:

1.  Across $\Sigma$, $U$ and $\dfrac{\partial U}{\partial n}$ are the same on both sides.
2.  Over the rest of $S_1$, $U = 0$ and $\dfrac{\partial U}{\partial n} = 0$.

These are not exactly true because of fringing at the boundary of the aperture.  These assumptions are valid as long as $\Sigma \gg \lambda$.

For optical wavelengths, $k \gg \dfrac{1}{r_{01}}$, so

$\frac{\partial G}{\partial n}
\approx
\cos(\hat{n}, \vec{r}_{01}) \, jk \frac{e^{jk r_{01}}}{r_{01}}$

The resulting equation is

$$U(P_0)
= \frac{1}{4\pi} \iint_\Sigma\frac{e^{jk r_{01}}}{r_{01}}
\left[\frac{\partial U}{\partial n}-jk U \cos(\hat{n}, \vec{r}_{01})\right] ds$$

$U$ is the electric field.  $U(P_o)$ is the field that is being calculated and $U$ in the integral is the source electric field.

This equation is called the Kirchhoff diffration integrtal.  It has some inconsistencies that arise from the requirement of both the field and its normal derivative


# Rayleigh-Sommerfeld Formulation

We start from the Helmholtz–Kirchhoff integral given by

$U(P_0)=\iint_{S}\left[G\,\frac{\partial U}{\partial n}-U\,\frac{\partial G}{\partial n}\right]\,ds'$

where $G$ is a chosen Green's function, $U$ the total field on the surface $S$, and the integration is over the closed surface.

Sommerfeld introduces Dirichlet and Neumann Green's functions defined with respect to mirror-source points symmetric about the aperture plane.  

<img src="sommerfeld.jpg" width="500">

The Dirichlet Green's function (Eq. 5.12) is

$G_D(\mathbf r;\mathbf r',\mathbf r'')=\frac{e^{-jkR_1}}{4\pi R_1}-\frac{e^{-jkR_2}}{4\pi R_2}$

On the source plane $\left( z'=0 \right)$, $G_D|_{z'=0}=0$.

The Neumann Green's function is

$G_N(\mathbf r;\mathbf r',\mathbf r'')=\frac{e^{-jkR_1}}{4\pi R_1}+\frac{e^{-jkR_2}}{4\pi R_2}$.

On the source plane $\left( z'=0 \right)$, $\frac{\partial G_D}{\partial n}=0$.

Using the Dirichlet Green's function in the Helmholtz–Kirchhoff integral (so that $G_D|_{z'=0}=0$ on the aperture plane) simplifies the integral because the $G\,\partial U/\partial n$ term vanishes on $z'=0$. Thus only the $-U\,\partial G/\partial n$ term remains:
$U_{RS1}(P_0)=\iint_{S_a}\left(-U_{\text{inc}}\frac{\partial G_D}{\partial n}\right)\Bigg|_{z'=0}\,dx'\,dy'$

Compute the normal derivative of $G_D$ at the aperture plane ($\hat n=-\hat z$, normal pointing into the screen). For a single free-space term
$\nabla' \left(\frac{e^{-jkR}}{4\pi R}\right)=\left(jk+\frac{1}{R}\right)\frac{e^{-jkR}}{4\pi R}\,\frac{\mathbf R}{R}$,
and the normal component yields a factor $\hat n\cdot\frac{\mathbf R}{R}=-\frac{z_0}{R}$ when evaluated at $z'=0$ (with the sign handled by the chosen normal). Taking the difference between the two mirror-source contributions in $G_D$ yields (after evaluating at $z'=0$ and using the symmetry that makes the second mirror term cancel at the plane) a single effective term:
$\frac{\partial G_D}{\partial n}\Big|_{z'=0}=-\left(jk+\frac{1}{R}\right)\frac{e^{-jkR}}{2\pi R}\frac{z_0}{R}.$

The factor $1/(2\pi)$ arises because the two mirror terms combine and the subtraction in $G_D$ changes the denominator from $4\pi$ to $2\pi$ after simplifying at $z'=0$ — see the detailed evaluation in the textbook.

Substitute the $\partial G_D/\partial n$ into the expression for $U_{RS1}$ to get

$U_{RS1}(P_0)=\iint_{S_a}\left(-U_{\text{inc}}\right)\left[-\left(jk+\frac{1}{R}\right)\frac{e^{-jkR}}{2\pi R}\frac{z_0}{R}\right]\,dx'\,dy'$

The two negatives cancel, giving

$U(P_o)=\iint_{S_a} U_{inc

}(x',y',0)\left(jk+\frac{1}{R}\right)\frac{e^{-jkR}}{2\pi R}\frac{z_0}{R}\,dx'\,dy'$

Recognize $\cos\theta=\dfrac{z_0}{R}$. Rewriting slightly we get 

$U(P_o)=\iint_{S_a} U_{inc}(x',y',0)\left(jk+\frac{1}{R}\right)\frac{e^{-jkR}}{2\pi R}\frac{z_0}{R}\,dx'\,dy'$ (See book Equation 5.18)

## Fresnel Approximation

The distance away from the aperture is typically much larger than the aperture size.  Factor the $z$ out of $r_{01}$

$r_{01} = z \sqrt{\left(\frac{x-x'}{z}\right)^2 + \left(\frac{y-y'}{z}\right)^2 + 1}$

Using binomial expansion
$(1+\Delta)^{1/2} \approx 1 + \frac{\Delta}{2} - \frac{\Delta^2}{8} + \dots$

The change of the $e^{jkr_{01}}$ term is much faster than the $\frac{1}{r^2}$ term.  

Inside exponential retain two terms 

$\exp \left(jkr_{01} \right) = \exp\left[jkz \sqrt{ \frac{(x-x')^2}{z^2} + \frac{(y-y')^2}{z^2} +1 } \right]$

$\exp \left(jkr_{01} \right) = \exp\left[jkz \left( 1+\frac{(x-x')^2}{2z^2} + \frac{(y-y')^2}{2z^2} \right) \right]$

$\exp \left(jkr_{01} \right) \approx \exp \left[jkz \right] \, \exp \left[ jkz \left( \frac{(x-x')^2}{2z^2} + \frac{(y-y')^2}{2z^2}\right) \right]$

$\exp \left(jkr_{01} \right) \approx \exp \left[jkz \right] \, \exp \left[ \frac{jk}{2z} \left( (x-x')^2 + (y-y')^2 \right) \right]$

The integration variables are $dx'$ and $dy'$; therefore, the term $\exp^{jkz}$ can be pulled iout of the integral.

Outside of the exponential retain a single term

$\frac{1}{r^2} \approx \frac{1}{z^2}$

$\frac{\exp \left(jk \sqrt{(x-x')^2 + (y-y')^2 + z^2} \right) }{(x-x')^2 + (y-y')^2 + z^2}
\approx
\frac{e^{jkz}}{z^2} \exp\left[j\frac{k}{2z}\left((x-x')^2 + (y-y')^2\right)\right]$

The third term in the binomial expansion is often used as the error approximation for the Fresnel approximation. The exponential function cycles every $\pi$.  Therefore, the error approximation for 

$\exp \left(jkr_{01} \right) \approx \exp \left[jkz \right] \, \exp \left[ \frac{jk}{2z} \left( (x-x')^2 + (y-y')^2 \right) \right]$

is

$kz \left( \frac{1}{8} \right) \left( \frac{(x-x')^2}{z^2}+\frac{(y-y')^2}{z^2} \right)^2  \ll 2 \pi$
